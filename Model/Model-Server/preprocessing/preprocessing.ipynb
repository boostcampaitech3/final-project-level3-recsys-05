```python
import numpy as np
import pandas as pd
```

```python
api_df = pd.read_csv('/opt/ml/level-3-final-project/Model/data/problem_api_1000_to_25150.csv')
meta_df = pd.read_csv('/opt/ml/level-3-final-project/Model/data/problem_meta_1000_to_25150.csv')
total_df = meta_df.merge(api_df, left_on='problem_num', right_on='problem_num').sort_values('problem_num').reset_index(drop = True)
total_df['problem_num'] = total_df['problem_num'].astype(str)
```

```python
problem_id2idx = {}

idx2problem_id = {}

for idx, problem_id in enumerate(total_df['problem_num'].unique()):
    problem_id2idx[problem_id] = idx

    idx2problem_id[idx] = problem_id
```

```python
import json

with open("problem_id2idx.json", "w") as json_file:
    json.dump(problem_id2idx, json_file)

with open("idx2problem_id.json", "w") as json_file:
    json.dump(idx2problem_id, json_file)
```

```python
with open('problem_id2idx.json', 'r', encoding = 'utf-8') as f:
    load_problem_id2idx= json.load(f)
```
