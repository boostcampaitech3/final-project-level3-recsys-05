성범님의 태그 preprocessing 코드 참고

```python
tag_list = []
def get_preprocessing_tags(tags):
    global tag_list
    tags = eval(tags)
    if tags:
        tags = [tag['key'] for tag in tags]
        tag_list += tags
    return tags

problem['preprocessing_tags'] = problem['problem_tags'].apply(lambda x : get_preprocessing_tags(x))
clean_problem['preprocessing_tags'] = clean_problem['problem_tags'].apply(lambda x: get_preprocessing_tags(x))
tag_list = list(set(tag_list))
```

```python
problem['preprocessing_tags'].value_counts()
```

```python
### problem.csv
tags = list(chain(*problem['preprocessing_tags'].to_list()))
tags_count = Counter(tags).most_common()
print("태그 개수 : ", len(set(tags)))
print("가장 많이 등장한 태그 및 카운트 : \n", tags_count[:10])
print("가장 적게 등장한 태그 및 카운트 : \n", tags_count[-10:])
```
